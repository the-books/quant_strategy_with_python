```python
import requests
import bs4
import pandas as pd
import time
```

```python
price_url = 'https://fchart.stock.naver.com/sise.nhn?symbol=051900&timeframe=day&count=500&requestType=0'
price_data = requests.get(price_url)
```

```python
price_data_bs = bs4.BeautifulSoup(price_data.text, 'html')
```

```python
# price_data_bs
```

```python
item_list = price_data_bs.findAll('item')
```

```python
# temp = item_list[0]['data']
```

```python
# temp.split('|')
```

```python
date_list = []
price_list = []

for item in item_list:
    temp_data = item['data']
    datas = temp_data.split('|')
    date_list.append(datas[0])
    price_list.append(datas[4])
    
price_df = pd.DataFrame({'종가':price_list}, index=date_list)
```
